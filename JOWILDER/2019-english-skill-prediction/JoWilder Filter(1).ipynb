## New Features
Create basic features from original data.

```python
# get all the features as time to next obj
times_to_next_obj = [f'obj{i}_otime_to_next_obj' for i in range(80)]
new_feat_df = LogDF.copy()
# times_in_objectives = sum all time in objectives
new_feat_df['times_in_objectives'] = new_feat_df[times_to_next_obj].sum(axis=1)
# levels_played = max-min+1
new_feat_df = new_feat_df.eval('levels_played = (max_level-start_level+1)')
new_feat_df['R1_quiz_response_bin'] = new_feat_df.loc[:,'R1_quiz_response'].replace({1:0,2:1})
new_feat_df['R2_quiz_response_bin'] = new_feat_df.loc[:,'R2_quiz_response'].replace({1:0,2:1, 3:1})
new_feat_meta = ['Added feat: times_in_objectives = sum(obj{i}_otime_to_next_obj for i in range(80))',
                 'Added feat: levels_played = (max_level-start_level+1)',
                 'Added feat: R1_quiz_response_bin = [R1_quiz_response].replace({1:0, 2:1})',
                 'Added feat: R2_quiz_response_bin = [R2_quiz_response].replace({1:0, 2:1, 3:1})']
```

Add onehotencoded features

```python
cat_list = jw_util.QA_1_cats+jw_util.onext_int_cats
cat_cols = jw_util.QA_1_feats+jw_util.onext_int_feats+[f'Q{i}_A2' for i in range(19)]+[f'Q{i}_A3' for i in range(19)]
onehot_df = pd.get_dummies(new_feat_df, columns=cat_cols)
onehot_df_meta = ['Expanded into onehot, ignoring NaN: '+', '.join(cat_cols)]
```

## Export

```python
def save_csv_and_meta(df, meta_list, save_dir, csv_name, meta_name=None, permissions='w+'):
    if csv_name.endswith(('.tsv', '.csv')):
        extension = csv_name[-4:]
        csv_name = csv_name[:-4]
    else:
        extension = '.csv'
    separator = '\t' if extension == '.tsv' else ','
    meta_name = meta_name or csv_name+ '_meta.txt'
    meta_text = 'Metadata:\n'+'\n'.join(meta_list)
    with open(os.path.join(save_dir, meta_name), permissions) as f:
        f.write(meta_text)
    with open(os.path.join(save_dir, csv_name)+extension, permissions) as f:
        for l in meta_text.splitlines():
            f.write(f'# {l}\n')
        f.write('\n')
        df.to_csv(f, sep=separator)

    return None, []
meta = import_meta + new_feat_meta + onehot_df_meta + filtered_df_meta
save_csv_and_meta(filtered_df, meta, os.path.join(PROJECT_DIR, 'Colab', 'Output Data'), 'JM_202006101204_all_woutNaN.csv')
```
