## New Features

#### Create basic features from original data.

```python
# get all the features as time to next obj
times_to_next_obj = [f'obj{i}_otime_to_next_obj' for i in range(80)]
new_feat_df = LogDF.copy()
# times_in_objectives = sum all time in objectives
new_feat_df['times_in_objectives'] = new_feat_df[times_to_next_obj].sum(axis=1)
# levels_played = max-min+1
new_feat_df = new_feat_df.eval('levels_played = (max_level-start_level+1)')
new_feat_df['R1_quiz_response_bin'] = new_feat_df.loc[:,'R1_quiz_response'].replace({1:0,2:1})
new_feat_df['R1_quiz_response_0v12'] = new_feat_df.loc[:,'R1_quiz_response'].replace({1:1,2:1})
new_feat_df['R1_quiz_response_01v2'] = new_feat_df.loc[:,'R1_quiz_response'].replace({1:0,2:1})
new_feat_df['R1_quiz_response_bin_x'] = new_feat_df.loc[:,'R1_quiz_response'].replace({1:np.nan,2:1})
new_feat_df['R2_quiz_response_bin'] = new_feat_df.loc[:,'R2_quiz_response'].replace({1:0,2:1, 3:1})
new_feat_df['R2_quiz_response_bin0v123'] = new_feat_df.loc[:,'R2_quiz_response'].replace({1:1,2:1, 3:1})
new_feat_df['R2_quiz_response_bin01v23'] = new_feat_df.loc[:,'R2_quiz_response'].replace({1:0,2:1, 3:1})
new_feat_df['R2_quiz_response_bin012v3'] = new_feat_df.loc[:,'R2_quiz_response'].replace({1:0,2:0, 3:1})
new_feat_df['R2_quiz_response_bin_x'] = new_feat_df.loc[:,'R2_quiz_response'].replace({1:np.nan,2:np.nan, 3:1})
new_feat_meta = ['Added feat: times_in_objectives = sum(obj{i}_otime_to_next_obj for i in range(80))',
                 'Added feat: levels_played = (max_level-start_level+1)',
                 'Added feat: R1_quiz_response_bin = [R1_quiz_response].replace({1:0, 2:1})',
                 'Added feat: R2_quiz_response_bin = [R2_quiz_response].replace({1:0, 2:1, 3:1})']
```

#### Sparse Interactions

```python
sparse_int_feats = []
int_prefs = []
num_encounters_columns = [c for c in reduced_col_df.columns if c.endswith("inum_enc")]
tot_count = len(reduced_col_df)
for c in num_encounters_columns:
  count = reduced_col_df[c].count()
  if count < abs_int_thresh:
    int_pref = c.split('_')[0]
    int_prefs.append(int(int_pref[3:]))
    percent = round(count/tot_count*100,2)
    reduced_col_df_meta.append(f"Remove {int_pref}: {count}/{tot_count} ({percent}%) occurance.")
    sparse_int_feats.extend([c2 for c2 in reduced_col_df.columns if c2.startswith(int_pref)])
reduced_col_df = reduced_col_df.drop(columns=sparse_int_feats)
reduced_col_df_meta.append(f'Removed {len(int_prefs)} other interactions with below {abs_int_thresh} ({int(SPARSE_INT_THRESH*100)}%) occurances: {sorted(int_prefs)}')
```

#### Sparse one-hots

```python
from collections import defaultdict
sparse_cats = defaultdict(lambda: [])
dflen = len(reduced_col_df)
y_cols, bool_cols, int_cols = jw_util.separate_columns(reduced_col_df, bool_dtype="uint8")
for bool_col in bool_cols:
    occurances = reduced_col_df[bool_col].sum()
    if occurances < abs_thresh:
        # print(bool_col, f"only has {occurances} occurances!")
        other_cat = '_'.join(bool_col.split('_')[:-1]+['other'])
        sparse_cats[other_cat].append(bool_col)
for cat, base_cats in sparse_cats.items():
    reduced_col_df.loc[:,cat] = reduced_col_df.loc[:,base_cats].any(axis=1).astype("uint8")
all_sparse_cats = [x for sublist in sparse_cats.values() for x in sublist]
reduced_col_df = reduced_col_df.drop(columns=all_sparse_cats)
reduced_col_df_meta.append(f'Grouped the following {len(all_sparse_cats)} categories into other: {all_sparse_cats}')
```

#### Remove any other sparse columns

```python
other_sparse = []
for col in reduced_col_df.columns:
    ct = reduced_col_df[col].count()
    if ct < abs_thresh:
        reduced_col_df_meta.append(f'Remove {col}: {ct} ({round(ct/len(reduced_col_df)*100,2)}%) occurances.')
        other_sparse.append(col)
reduced_col_df = reduced_col_df.drop(columns=other_sparse)
```

```python
#### out_fname_prefix

out_fname_prefix = 'JM_202007061210'

#### Objective Min/Max
# Set sess_end_obj_min and sess_end_obj_max (inclusive). 
# Set None for no min/max. 

# END_FEAT_OBJ identifies the objective after which subsequent features should not be included.

# Assessment 1 ends at 11.

# Assessment 2 ends at 42.

# Assessment 3 ends at 75.

# (use 12, 43, 76 to make sure players all finished the objective)
 
# See https://docs.google.com/spreadsheets/d/1-gN6YPPq51B7DWEaN-M-PVYsHfpba7pFTi9ciDQurjI/edit#gid=1825984699 for details.
# It contains sheet "Game Segments" with a map of objectives to levels, checkpoints, questions, chapters, and assessments.

SESS_END_OBJ_MIN = 76
SESS_END_OBJ_MAX = None


#### Drop Cols

DROP_COLS = [f'Q{q}_answers' for q in range(19)] + ["play_year", "play_month", "play_day", "play_hour", "play_minute",
                                                    "play_second",
                                                    "_continue", "continue", "save_code", "music", "hq", "fullscreen",
                                                    "persistentSessionID", ]
# DROP_COLS.extend([f'Q{q}_A{a}' for q in range(19) for a in [2,3]])
# DROP_COLS.extend([f'Q{q}_A{a}_time' for q in range(19) for a in [2,3]])

#### Sparse Thresh
# This is the minimum percent True allowed for onehot arrays before grouped into other. This is also the minimum percent required to keep any given interaction feature

for thresh in [0, .01, .05, .10]:
    SPARSE_INT_THRESH = thresh
    SPARSE_THRESH = thresh
    for ONLY_NONNATIVE in [True, False]:
        DROP_UNREACHED_FEATS = True
        for DROP_UNREACHED_FEATS in [True, False]:
            for SESS_END_OBJ_MIN in [12, 43, 76]:
                ### Processing
                END_FEAT_OBJ = SESS_END_OBJ_MIN - 1
                REMOVE_INTS = list(range(END_FEAT_OBJ+1, 80))
                _end_obj_to_end_opt_int = {
                    11: 113,
                    42: 167,
                    75: 185
                }
                REMOVE_INTS.extend(list(range(_end_obj_to_end_opt_int[END_FEAT_OBJ], 189)))
                SESS_END_Q = jw_util.end_obj_to_last_Q[END_FEAT_OBJ]
                SESS_END_LVL = jw_util.end_obj_to_last_lvl[END_FEAT_OBJ]
                print("Last Q:", SESS_END_Q, "Last LVL:", SESS_END_LVL)


                ## Objective Filter
                sess_end_obj_filters = []
                if SESS_END_OBJ_MIN is not None:
                    sess_end_obj_filters.append(f"sess_end_obj>{SESS_END_OBJ_MIN -1}")
                if SESS_END_OBJ_MAX is not None:
                    sess_end_obj_filters.append(f"sess_end_obj<{SESS_END_OBJ_MAX+1}")
                if ONLY_NONNATIVE:
                    sess_end_obj_filters.append("R0_quiz_response == 1")
                objective_filtered_df, objective_filtered_df_meta = feat_util.filter_df(filtered_df, sess_end_obj_filters, fillna=None);
                objective_filtered_df_meta

                ## Remove Columns

                reduced_col_df = objective_filtered_df.copy()
                abs_thresh = int(SPARSE_THRESH*len(reduced_col_df))
                abs_int_thresh = int(SPARSE_INT_THRESH*len(reduced_col_df))
                reduced_col_df_meta = [
                    f'*arg*: SPARSE_THRESH = {SPARSE_THRESH}',
                    f'({abs_thresh}/{len(reduced_col_df)} sessions)',
                    f'*arg*: SPARSE_INT_THRESH = {SPARSE_INT_THRESH}',
                    f'({abs_int_thresh}/{len(reduced_col_df)} sessions)',
                    f'*arg*: SESS_END_Q = {SESS_END_Q}',
                    f'*arg*: SESS_END_LVL = {SESS_END_LVL}',
                    f'*arg*: END_FEAT_OBJ = {END_FEAT_OBJ}',
                    f'*arg*: DROP_COLS = {DROP_COLS}',

                ]
                reduced_col_df_meta

                #### Remove drop_cols

                _drop_cols = [c for c in DROP_COLS if c in reduced_col_df.columns]
                reduced_col_df = reduced_col_df.drop(columns = _drop_cols)
                reduced_col_df_meta.append(f'Removed these columns from DROP_COLS: {_drop_cols}')

                #### Unreached objs, qs, lvls, ints

                if DROP_UNREACHED_FEATS:
                    drop_obj_feats = []
                    for obj in range(END_FEAT_OBJ+1, 80):
                        prefix = f'obj{obj}'
                        drop_obj_feats.extend([c2 for c2 in reduced_col_df.columns if c2.startswith(prefix)])
                    reduced_col_df = reduced_col_df.drop(columns=drop_obj_feats)
                    reduced_col_df_meta.append(f'Removed {len(drop_obj_feats)} objective features range obj{END_FEAT_OBJ+1}-79.')

                    drop_q_feats = []
                    for q in range(SESS_END_Q+1,19):
                        prefix = f'Q{q}'
                        drop_q_feats.extend([c2 for c2 in reduced_col_df.columns if c2.startswith(prefix)])
                    reduced_col_df = reduced_col_df.drop(columns=drop_q_feats)
                    reduced_col_df_meta.append(f'Removed {len(drop_q_feats)} question features range q{SESS_END_Q+1}-18.')

                    drop_lvl_feats = []
                    for lvl in range(SESS_END_LVL+1,24):
                        prefix = f'lvl{lvl}'
                        drop_lvl_feats.extend([c2 for c2 in reduced_col_df.columns if c2.startswith(prefix)])
                    reduced_col_df = reduced_col_df.drop(columns=drop_lvl_feats)
                    reduced_col_df_meta.append(f'Removed {len(drop_lvl_feats)} level features range lvl{SESS_END_LVL+1}-23.')

                    drop_int_feats = []
                    for int_enum in REMOVE_INTS:
                        prefix = f'int{int_enum}'
                        drop_int_feats.extend([c2 for c2 in reduced_col_df.columns if c2.startswith(prefix)])
                    reduced_col_df = reduced_col_df.drop(columns=drop_int_feats)
                    reduced_col_df_meta.append(f'Removed {len(REMOVE_INTS)} interactions from REMOVE_INTS: {sorted(REMOVE_INTS)}.')

                #### Sparse Interactions

                sparse_int_feats = []
                int_prefs = []
                num_encounters_columns = [c for c in reduced_col_df.columns if c.endswith("inum_enc")]
                tot_count = len(reduced_col_df)
                for c in num_encounters_columns:
                    count = reduced_col_df[c].count()
                    if count < abs_int_thresh:
                        int_pref = c.split('_')[0]
                        int_prefs.append(int(int_pref[3:]))
                        percent = round(count/tot_count*100,2)
                        reduced_col_df_meta.append(f"Remove {int_pref}: {count}/{tot_count} ({percent}%) occurance.")
                        sparse_int_feats.extend([c2 for c2 in reduced_col_df.columns if c2.startswith(int_pref)])
                reduced_col_df = reduced_col_df.drop(columns=sparse_int_feats)
                reduced_col_df_meta.append(f'Removed {len(int_prefs)} other interactions with below {abs_int_thresh} ({int(SPARSE_INT_THRESH*100)}%) occurances: {sorted(int_prefs)}')

                #### Sparse one-hots

                from collections import defaultdict
                sparse_cats = defaultdict(lambda: [])
                dflen = len(reduced_col_df)
                y_cols, bool_cols, int_cols = jw_util.separate_columns(reduced_col_df, bool_dtype="uint8")
                for bool_col in bool_cols:
                    occurances = reduced_col_df[bool_col].sum()
                    if occurances < abs_thresh:
                        # print(bool_col, f"only has {occurances} occurances!")
                        other_cat = '_'.join(bool_col.split('_')[:-1]+['other'])
                        sparse_cats[other_cat].append(bool_col)
                for cat, base_cats in sparse_cats.items():
                    reduced_col_df.loc[:,cat] = reduced_col_df.loc[:,base_cats].any(axis=1).astype("uint8")
                all_sparse_cats = [x for sublist in sparse_cats.values() for x in sublist]
                reduced_col_df = reduced_col_df.drop(columns=all_sparse_cats)
                reduced_col_df_meta.append(f'Grouped the following {len(all_sparse_cats)} categories into other: {all_sparse_cats}')

                #### Remove any other sparse columns

                other_sparse = []
                for col in reduced_col_df.columns:
                    ct = reduced_col_df[col].count()
                    if ct < abs_thresh:
                        reduced_col_df_meta.append(f'Remove {col}: {ct} ({round(ct/len(reduced_col_df)*100,2)}%) occurances.')
                        other_sparse.append(col)
                reduced_col_df = reduced_col_df.drop(columns=other_sparse)

                #### Inspect reduced_col_df and meta

                print(*reduced_col_df_meta, sep='\n')
                reduced_col_df



                os.getcwd()

                ## Export

                obj_str = f'obj{SESS_END_OBJ_MIN or ""}-{SESS_END_OBJ_MAX or ""}'
                thresh_str = f'thresh{int(SPARSE_INT_THRESH*100)}_{int(SPARSE_THRESH*100)}'
                unreached_str = f'_withUnreached' if not DROP_UNREACHED_FEATS else ''
                nonnative_str = f'_onlyNonnative' if ONLY_NONNATIVE else ''
                print(unreached_str, nonnative_str, obj_str, thresh_str)

                meta = import_meta + new_feat_meta + onehot_df_meta + filtered_df_meta +\
                    objective_filtered_df_meta + reduced_col_df_meta
                outdir = os.path.join(PROJECT_DIR, 'Colab', 'Output Data')
                out_fname = f'{out_fname_prefix}{unreached_str}{nonnative_str}_{obj_str}_{thresh_str}.csv'
                feat_util.save_csv_and_meta(reduced_col_df, meta, outdir, out_fname)

                # from datetime import datetime
                # date_time = datetime.now().strftime("%Y%m%d%H%M")
                # print("date and time:",date_time)

                ## View Final Meta

                print("Output to:", outdir)
                print("File name:", out_fname)

                print(*meta, sep='\n')
```
